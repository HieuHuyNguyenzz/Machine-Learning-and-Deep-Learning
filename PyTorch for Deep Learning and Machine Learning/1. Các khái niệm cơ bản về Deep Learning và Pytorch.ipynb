# Deep Learning




## 1. Deep Learning là gì?
Một số khái niệm cơ bản
 - Machine Learning là một phương pháp chuyển đổi một số 'thing' (dữ liệu) thành các chữ số và tìm điểm chung hay các mẫu chung giữa các điểm dữ liệu này

 - Dữ liệu ở đây có thể dưới nhiều dạng như ảnh, văn bản, bảng, số, video, ... miễn là có thể chuyển đổi sang dạng số và tìm mẫu chung của các số này.

 - Việc tìm kiếm các mẫu chung này được thực hiện sử dụng các thuật toán
 Machine Learning, các thuật toán này dựa trên lập trình và toán học.

 - Deep Learning là một tập con các thuật toán trong Machine Learning, trong đó dữ liệu sẽ được đưa vào các mô hình mạng Neural nhiều lớp (sâu).



## 2. Tại sao lại sử dụng Machine Learning và Deep Learning
Machine Learning và Deep Learning phù hợp trong các trường hợp sau:
- Đối với các vấn đề có độ phức tạp lớn, việc lập trình các luật để tạo thành một thuật toán là gần như không thể -> ML và DL có khả năng tự khám phá ra các luật này.
- Cần một thuật toán hay mô hình có khả năng thích ứng cũng như học từ các bối cảnh, tri thức mới một cách dễ dàng.
- Nhu cầu khám phá và học các tri thức từ một bộ dữ liệu rất lớn

Machine Learning và Deep Learning có thể không thực sự phù hợp trong các trường hợp:
- Khi cần chứng mình tính đúng đắn
- Khi các phương pháp truyền thống tỏ ra tốt hơn
- Khi việc xảy ra lỗi là không chấp nhận được
- Khi dữ liệu là không nhiều



## 3. Neural Networks
Neural Networks là một loại mạng được thiết kế để học được từ các loại dữ liệu phi cấu trúc một cách hiệu quả. Dựa trên cảm hứng từ cấu trúc của Neural thần kinh, một mô hình Neural Network gồm 3 lớp: Lớp đầu vào (Input Layer), lớp ẩn (Hidden Layer) và lớp đầu ra (Output Layer). Mỗi lớp gồm nhiều các nút chứa tham số.

- Lớp đầu vào có tác dụng đọc dữ liệu đầu vào
- Lớp ẩn trích xuất các đặc trưng từ lớp đầu vào
- Lớp đầu ra sử dụng các đặc trưng từ lớp ẩn và chuyển đổi thành đầu ra của mô hình dưới dạng xác suất


## 4. Các dạng 'Học'
### Supervised Learning
Supervised Learning (Học có giám sát) là một phương pháp trong Machine Learning, nơi mô hình được huấn luyện trên một tập dữ liệu có nhãn. Mỗi mẫu dữ liệu bao gồm đầu vào (input) và đầu ra mong muốn (output hoặc nhãn). Mục tiêu là học một hàm ánh xạ từ đầu vào đến đầu ra, sao cho mô hình có thể dự đoán chính xác đầu ra cho các dữ liệu mới chưa từng thấy.

###Unsupervised và Self-supervised Learning
- Unsupervised Learning là một phương pháp trong Machine Learning, nơi mô hình được huấn luyện trên dữ liệu không có nhãn. Mục tiêu là khám phá cấu trúc ẩn hoặc mẫu hình trong dữ liệu mà không cần biết trước đầu ra mong muốn.
- Self-Supervised Learning là một phương pháp kết hợp giữa Supervised và Unsupervised Learning. Mô hình tự tạo ra nhãn từ dữ liệu thô mà không cần sự can thiệp của con người, sau đó sử dụng các nhãn này để huấn luyện.

### Transfer Learning
Transfer Learning là một kỹ thuật trong Machine Learning, nơi một mô hình đã được huấn luyện trên một bài toán hoặc tập dữ liệu cụ thể được tái sử dụng để giải quyết một bài toán khác có liên quan. Thay vì huấn luyện mô hình từ đầu, Transfer Learning tận dụng kiến thức đã học được từ bài toán trước đó để cải thiện hiệu suất và giảm thời gian huấn luyện.

### Reinforcement Learning
Reinforcement Learning (RL) là một phương pháp trong Machine Learning, nơi một agent học cách thực hiện các hành động trong một môi trường để tối đa hóa một phần thưởng (reward) tích lũy. Khác với Supervised Learning và Unsupervised Learning, RL không dựa trên dữ liệu có nhãn hay cấu trúc ẩn mà dựa trên sự tương tác liên tục giữa agent và môi trường.

# Pytorch


## 1. Pytorch là gì?
Pytorch là một thư viện mã nguồn mở được thiết kể bởi Meta để cài đặt và triển khai các phương pháp và thuật toán Deep Learning sử dụng Python.

Một số các ưu điểm của Pytorch:
- Cho phép chạy trên một hoặc nhiều GPU với tốc độ cao
- Cung cấp một lượng lớn các mô hình Deep Learning có sẵn
- Cung cấp full-stack các công cụ cần thiết để triển khai mô hình deep learning như tiền xử lý dữ liệu, mô hình hóa dữ liệu, huấn luyện và triển khai AI trên các ứng dụng hoặc đám mây.

Pytorch là thư viện được sử dụng nhiều nhất trong các nghiên cứu về lĩnh vực deep learning.

## 2. Tensor là gì?
Tensor là một khái niệm toán học và là một cấu trúc dữ liệu cơ bản trong các thư viện Machine Learning. Tensor có thể được coi là một mảng đa chiều (multi-dimensional array) tổng quát hóa của các khái niệm như scalar, vector, và matrix.

Các loại Tensor cơ bản
Scalar (0D Tensor):

- Một số đơn lẻ, không có chiều.
- Ví dụ: 5, 3.14.

Vector (1D Tensor):
- Một mảng một chiều gồm các số.
- Ví dụ: [1, 2, 3, 4].

Matrix (2D Tensor):
- Một mảng hai chiều gồm các số.

- Ví dụ: [[1, 2, 3], [4, 5, 6]]
